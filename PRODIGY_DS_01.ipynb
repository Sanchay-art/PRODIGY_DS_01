```python
# Importing Required Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Generating the dataset

```python
# Set random seed for reproducibility
np.random.seed(42)

# Generate sample dataset
n = 1000  # Number of individuals

# Continuous variables
age = np.random.randint(18, 80, size=n)  # Age between 18 and 80
income = np.random.randint(10000, 200000, size=n)  # Income in INR

# Categorical variables
gender = np.random.choice(['Male', 'Female', 'Other'], size=n, p=[0.48, 0.48, 0.04])
state = np.random.choice(['Uttar Pradesh', 'Maharashtra', 'Bihar', 'West Bengal', 'Tamil Nadu', 'Karnataka', 'Gujarat'], size=n)
education = np.random.choice(['Primary', 'Secondary', 'Higher Secondary', 'Graduate', 'Postgraduate'], size=n)
occupation = np.random.choice(['Farmer', 'Teacher', 'Engineer', 'Doctor', 'Student', 'Businessperson', 'Unemployed'], size=n)
```

### Displaying the Data 

```python
# Create the dataframe
data = pd.DataFrame({
    'Age': age,
    'Income': income,
    'Gender': gender,
    'State': state,
    'Education': education,
    'Occupation': occupation
})

# Displaying first few rows of the dataset
data.head()
```

### Bar Chart and Histogram

```python
# Plotting
fig, axes = plt.subplots(2, 1, figsize=(12, 12))

# Bar chart for categorical variable (Gender)
sns.countplot(x='Gender', data=data, palette='pastel', ax=axes[0])
axes[0].set_title('Distribution of Gender')
axes[0].set_xlabel('Gender')
axes[0].set_ylabel('Count')

# Histogram for continuous variable (Age)
sns.histplot(data['Age'], bins=15, kde=True, color='skyblue', ax=axes[1])
axes[1].set_title('Distribution of Age')
axes[1].set_xlabel('Age')
axes[1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```
